### Find mean weights of species 'DM', for year 2000, 01 and 02

```python
temp_df = surveys_df[surveys_df.year >= 2000]
temp_df[temp_df.species_id == 'DM'].mean()['weight']
```

```python
# what about mean for year 2000, mean for 2001 and mean for 2002
temp_df[temp_df.species_id == 'DM'].groupby('year').mean()['weight']
```
