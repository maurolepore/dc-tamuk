# Visualization
matplotlib
plotnine package to mimick ggplot in R

```python
import plotnine as p9
```

```python
surveys_complete = surveys_df.dropna()
```

```python
%matplotlib inline
```

```python
p9.ggplot(data=surveys_complete,
         mapping=p9.aes(x='weight', y='hindfoot_length')) + p9.geom_point()
```

```python
surveys_plot = p9.ggplot(data=surveys_complete,
         mapping=p9.aes(x='weight', y='hindfoot_length'))

surveys_plot + p9.geom_point()
```

### Bar plot of numbers of record per plot_id using surveys_complete (Hints: you only need to specify x, no y. And you don't need to do anything pandas)

```python
p9.ggplot(data=surveys_complete,
         mapping=p9.aes(x='plot_id')) + p9.geom_bar()
```

```python
surveys_plot + p9.geom_point(alpha = 0.1)
```

```python
surveys_plot + p9.geom_point(alpha = 0.1, color = 'blue')
```

```python
surveys_plot = p9.ggplot(data=surveys_complete,
         mapping=p9.aes(x='weight', y='hindfoot_length', color = 'species_id'))

surveys_plot + p9.geom_point(alpha=0.1)
```

```python
surveys_plot = p9.ggplot(data=surveys_complete,
         mapping=p9.aes(x='weight', y='hindfoot_length', color = 'species_id'))

surveys_plot + p9.geom_point(alpha=0.1) + p9.facet_wrap('~ sex')
```
